```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Types of Graphs
1. plot -               line graph, between two parameters
2. bar -                bar graph of single parameter (optional cat_data)
3. hist -               histogram of a bunch of data i.e. frequency of single parameter
4. scatter -            dot graph between two params
5. boxplot / box -      statsitical overview in form of box for a single param
                        mean, outliers, qunatiles (25%, 75%), IQR
                        (optional - categotical data)
6. stem -               more like bar graph, Lines instead of bars 
7. violinplot -         Violin shaped bar graph statistical overview graph
                        spread of data (more like hist)
                        single parameter (optional cat-data)

```python
x = np.linspace(0,5,11)
x1 = np.arange(1,11,3)
```

```python
x
```

```python
y = x**3
```

```python
plt.plot(x,y)
# plt.show()
```

```python
plt.bar(x,height=y)
plt.title('Bar Graph')
```

```python
plt.hist(x)
plt.title('Histogram')
```

```python
plt.scatter(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.boxplot(y)
```

```python
plt.stem(y)
```

```python
plt.violinplot(y)
```

```python
y2=x**4
plt.plot(x,y,x,y2)
plt.legend(['y=x**3','y2=x**4'])
# plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(x,y, lw=3, marker='o',markersize=15,markerfacecolor='yellow')
plt.plot(x,y2,marker='v',markersize=10)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```
